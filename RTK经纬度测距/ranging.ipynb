```python
import pandas as pd
import csv
import math
from matplotlib import pyplot as plt
```

#### 将txt转为csv

```python
def convert_data(path1, path2):
    data = open(file=path1, mode='w', newline='')
    writer = csv.writer(data, dialect='excel')
    f = open(file=path2, mode='r')
    for line in f.readlines():
        line = line.replace(',', '\t')
        list_line = line.split()
        writer.writerow(list_line)
```

#### 数据清洗

```python
def clean_data1(path):
    # 读取数据
    df_raw = pd.read_csv(path, error_bad_lines=False)

    # 为数据重新设置列名
    df_raw.columns = ['name', 'time', 'north latitude', '列名4', 'East longitude', '列名6',
                      '列名7', '列名8', '列名9', '列名10', '列名11', '列名12', '列名13', '列名14']
    # 删除不必要的列
    df_remcol = df_raw.drop(labels=['列名4', '列名6', '列名7', '列名8', '列名9', '列名10',
                                    '列名11', '列名12', '列名13', '列名14'], axis=1)

    # 统计表格中各数据的数量
    counts = df_remcol["name"].value_counts()
    print(counts)

    # 只保留表格中GGA数据
    df_remcol = df_remcol[df_remcol.name.isin(['$GNGGA'])]

    # 重新设置表格索引
    df_new1 =  df_remcol.reset_index(drop=True)
    return df_new1
```

#### 提取两个表格中时间相同的元素

```python
def merge_data(df1, df2):
    df = pd.merge(df1, df2, how='inner', on=['name'])
    df.sort_index()
    return df
```

```python
df_new = merge_data(df1_new, df2_new)
# df_new
```

#### 测距

```python
Earth_Radius = 6378137

def sin_square(a):
    result = math.sin(a)
    return result**2

def rad(ang):
    return math.pi * ang / 180.0

list_distance = []
for i in range (df_new.shape[0]):
    lata, lnga = float(df_new.loc[i, 'north latitude_x']) / 100,\
                 float(df_new.loc[i, 'East longitude_x']) / 100

    latb, lngb = float(df_new.loc[i, 'north latitude_y']) / 100,\
                 float(df_new.loc[i, 'East longitude_y']) / 100
    # print(lata, lnga)
    # print(latb, lngb)
    # print('************')
    radlata, radlnga = rad(lata), rad(lnga)
    radlatb, radlngb = rad(latb), rad(lngb)
    # print(radlata, radlnga)
    #纬度差
    minus_a = radlata-radlatb
    #经度差
    minus_b = radlnga-radlngb
    distance = 2 * math.asin(math.sqrt(sin_square(minus_a / 2) + math.cos(radlata) * math.cos(radlatb) * sin_square(minus_b / 2))) * Earth_Radius
    list_distance.append(distance)
print(list_distance)
print('数据数量: ',len(list_distance))
```

#### 画图

```python
plt.plot(list_distance, linestyle = 'dotted')
plt.xlabel("counts")
plt.ylabel("distance")
```
